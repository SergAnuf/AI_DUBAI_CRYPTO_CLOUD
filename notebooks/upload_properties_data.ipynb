```python
def clean_property_features_to_json(raw):
    if pd.isna(raw) or str(raw).strip() in ["[]", ""]:
        return "[]"

    txt = str(raw).replace("\n", " ").strip()

    # Remove enclosing brackets and quotes artifacts
    txt = re.sub(r'[\[\]]', '', txt)

    # Ensure commas between adjacent quoted parts like: 'A' 'B' â†’ 'A','B'
    txt = re.sub(r"'\s*'", "', '", txt)

    # If ends with an open quote (e.g. 'Garage), close it
    if txt.count("'") % 2 != 0:
        txt += "'"

    # Remove redundant double quotes
    txt = txt.replace('"', '')

    # Split by single quotes and clean
    parts = [p.strip(" ,") for p in txt.split("'") if p.strip() and not p.strip().startswith(",")]

    # Normalize capitalization and remove duplicates
    cleaned = []
    for p in parts:
        if p and p not in cleaned:
            cleaned.append(p.strip().title())

    # Return valid JSON string list
    return json.dumps(cleaned, ensure_ascii=False)


```

```python
df["property_features"] = df["property_features"].apply(clean_property_features_to_json)
```
